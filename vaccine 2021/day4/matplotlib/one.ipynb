```python
import numpy as np

```

```python
items = ['software', 'fuel', 'food', 'shoes', 'steel']
quantity = [82, 88, 78, 74, 80]
expected = [90, 92, 86, 86, 86]

i = np.arange(5)
w = 0.3
```

```python
plt.bar(i, expected, w, color='green', label="target exports")
plt.bar(i+w, quantity, w, color='blue', label="achieved exports")

plt.ylabel("millions")
plt.xlabel("items")

plt.xticks(i+w/2, items)

plt.legend()
plt.show()
```

```python
plt.bar(i, expected, w, color='green', label="target exports")
plt.bar(i, quantity, w, color='blue', label="achieved exports", bottom=expected)

plt.ylabel("millions")
plt.xlabel("items")

plt.xticks(i, items)

plt.legend()
plt.show()
```
