```python
class AppError(Exception):
    """base class for all exceptions on my tool"""
    pass
```

```python
val = 50
```

```python
class AppError(Exception):
    """base class for all exceptions on my tool"""
    pass

class SmallValue(AppError):
    """the value is too small"""
    pass

class LargeValue(AppError):
    """the value is too large"""
    pass
```

```python
while True:
    try:
        num = int(input("\nenter the number:\n"))
        if num < val:
            raise SmallValue
        elif num > val:
            raise LargeValue
        else:
            print("you guessed it right")
            break
    except SmallValue:
        print(SmallValue.__doc__)
    except LargeValue:
        print(LargeValue.__doc__)
```
